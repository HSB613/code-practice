## **Part 2. Missing Value**



##### **ÏûêÏÇ∞Ï¥ùÍ≥ÑÏùò Í≤∞Ï∏°ÏπòÎ•º Ï≤òÎ¶¨ÌïòÏÑ∏Ïöî.**
ÏïûÏÑ† Î¨∏Ï†úÏóêÏÑúÎäî Í≤∞Ï∏°ÏπòÎ•º Îã®ÏàúÌûà 0ÏúºÎ°ú Ï≤òÎ¶¨ÌïòÏòÄÏúºÎÇò, Í≤∞Ï∏°ÏπòÎ•º Ï≤òÎ¶¨ÌïòÎäî Î∞©Î≤ïÏùÄ Í∑∏ Ïô∏ÏóêÎèÑ Ïó¨Îü¨Í∞ÄÏßÄÍ∞Ä ÏûàÏäµÎãàÎã§. <br>
Îã§ÏñëÌïú imputation Î∞©Î≤ïÏóê ÎåÄÌï¥ ÏïåÏïÑÎ≥¥Í≥†, Í∑∏ Ï§ë mean imputation Î∞©Î≤ïÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ `ÏûêÏÇ∞Ï¥ùÍ≥Ñ`Ïùò Í≤∞Ï∏°ÏπòÎ•º Ï≤òÎ¶¨Ìï¥Î≥¥ÏÑ∏Ïöî.

```python
import pandas as pd
import numpy as np

df1 = pd.read_excel("https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx", sheet_name = "008770 ")
df2 = pd.read_excel("https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx", sheet_name = "035250")

df1 = df1.set_index('Ï£ºÏöîÏû¨Î¨¥Ï†ïÎ≥¥')
df1 = df1.T
df1 = df1[1:]

df2 = df2.set_index('Ï£ºÏöîÏû¨Î¨¥Ï†ïÎ≥¥')
df2 = df2.T
df2 = df2[1:]

final_df1 = df1.fillna(df1.mean())
final_df2 = df2.fillna(df1.mean())

final_df1.isnull().sum()
```

## **Part 3.Feature Engineering**


##### **`Î∂ÄÏ±Ñ ÎπÑÏú® 2` ÎùºÎäî ÏÉàÎ°úÏö¥ ÌîºÏ≥êÎ•º ÏÉùÏÑ±ÌïòÏÑ∏Ïöî.**

> Î∂ÄÏ±Ñ ÎπÑÏú®(%) = (Î∂ÄÏ±ÑÏ¥ùÏï° / ÏûêÎ≥∏Ï¥ùÏï°) x 100

ÏúÑÏùò Í≥µÏãùÏùÑ ÌôúÏö©ÌïòÏó¨ `Î∂ÄÏ±Ñ ÎπÑÏú® 2` ÎùºÎäî columnÏùÑ ÏÉùÏÑ±Ìï©ÎãàÎã§. Í∑∏Î¶¨Í≥†, ÏÉùÏÑ±Ìïú columnÍ≥º Í∏∞Ï°¥ `Î∂ÄÏ±Ñ ÎπÑÏú®` columnÏùò Í∞íÏù¥ ÎèôÏùºÌïúÏßÄ ÌôïÏù∏Ìï¥Î¥ÖÎãàÎã§.

```python
final_df1['Î∂ÄÏ±Ñ ÎπÑÏú®']= final_df1['Î∂ÄÏ±ÑÏ¥ùÍ≥Ñ'] / final_df1['ÏûêÎ≥∏Ï¥ùÍ≥Ñ'] * 100

final_df1.sample(3)
```

##### **`Ï¢ÖÎ™©` Ïù¥ÎùºÎäî ÏÉàÎ°úÏö¥ ÌîºÏ≥êÎ•º ÏÉùÏÑ±ÌïòÏÑ∏Ïöî.**

**`.repeat()`**Î•º ÌôúÏö©ÌïòÏó¨ ÏÉàÎ°úÏö¥ `Ï¢ÖÎ™©`Ïù¥ÎùºÎäî ÏÉàÎ°úÏö¥ columnÏùÑ ÏÉùÏÑ±Ìï©ÎãàÎã§.
  - `df1`Ïùò `Ï¢ÖÎ™©` columnÏóêÎäî `008770` Î¨∏ÏûêÏó¥Ïù¥ Îã¥Í∏∞ÎèÑÎ°ù Ìï©ÎãàÎã§.  
  - `df2`Ïùò `Ï¢ÖÎ™©` columnÏóêÎäî `035250` Î¨∏ÏûêÏó¥Ïù¥ Îã¥Í∏∞ÎèÑÎ°ù Ìï©ÎãàÎã§. 

Í∞ÅÍ∞ÅÏùò Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê `Ï¢ÖÎ™©`Ïù¥ÎùºÎäî columnÏùÑ ÎßåÎì† ÌõÑ, ÏµúÏ¢ÖÏ†ÅÏúºÎ°ú Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ `df1`Í≥º `df2`Î•º **`.append()`**Î•º ÌôúÏö©ÌïòÏó¨ Ìï©Ï≥êÏ£ºÎèÑÎ°ù Ìï©ÎãàÎã§.

```python
final_df1['Ï¢ÖÎ™©'] = np.repeat('008770', len(final_df1))
final_df2['Ï¢ÖÎ™©'] = np.repeat('035250', len(final_df2))

add_df = final_df1.append(final_df2, ignore_index = True)

add_df.head(10)
```
